```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
import numpy as np
register_matplotlib_converters()

# Import data (Make sure to parse dates. Consider setting index column to 'date'.)
df = pd.read_csv('fcc-forum-pageviews.csv', index_col = 'date', parse_dates = ["date"])

# Clean data
df_clean = df.loc[
    (df['value'] >= df['value'].quantile(0.025))
    & (df['value'] <= df['value'].quantile(0.975))
]


def draw_line_plot():
    # Draw line plot
    fig = plt.figure(figsize=(14, 6))
    plt.plot(df_clean.index, df_clean.values)
    plt.title('Daily freeCodeCamp Forum Page Views 5/2016-12/2019')
    plt.xlabel('Date')
    plt.ylabel('Page Views')
    # Save image and return fig (don't change this part)
    fig.savefig('line_plot.png')
    return fig

def draw_bar_plot():
    # Copy and modify data for monthly bar plot
    fig = plt.figure(figsize = (30,15))
    df_box = df_clean.copy()
    df_box.reset_index(inplace=True)
    df_box['date'] = pd.to_datetime(df_box['date'], format = '%Y-%m-%d')
    df_box['year'] = [d.year for d in df_box.date]
    df_box['month'] = [d.strftime('%b') for d in df_box.date]
    df_box = df_box.drop('date', axis = 1)
    #df_bar = None

    # Draw bar plot
    mi_sum = df_box.groupby(by = ['year', 'month']).mean()
    df_sum = mi_sum.index.to_frame(index = False)
    df_sum['value'] = mi_sum['value'].values

    sort_order = ['Jan','Feb','Mar','Apr',
              'May','Jun','Jul','Aug',
              'Sep','Oct','Nov','Dec'
             ]
    df_sum.index = pd.CategoricalIndex(df_sum['month'], categories = sort_order, ordered = True)
    df_sum.sort_index()
    df_sum = df_sum.sort_index().reset_index(drop = True)

    sns.catplot(x="month", y = 'value',col="year",
                data = df_sum,
                kind="bar")
    
    # Save image and return fig (don't change this part)
    fig.savefig('bar_plot.png')
    return fig

def draw_box_plot():
    fig = plt.figure(figsize = (30,10))
    ax1 = plt.subplot(1,2,1)
    ax2 = plt.subplot(1,2,2)
    # Prepare data for box plots (this part is done!)
    df_box = df_clean.copy()
    df_box.reset_index(inplace=True)
    df_box['year'] = [d.year for d in df_box.date]
    df_box['month'] = [d.strftime('%b') for d in df_box.date]
    #df_month_group = df_box.groupby(['month']).sum()
    _ = sns.boxplot(x = df_box['year'] ,y =  df_box['value'], ax = ax1)
    sort_order = ['Jan','Feb','Mar','Apr',
                  'May','Jun','Jul','Aug',
                  'Sep','Oct','Nov','Dec'
                 ]
    df_box.index = pd.CategoricalIndex(df_box['month'], categories = sort_order, ordered = True)
    df_box = df_box.sort_index()
    _ = sns.boxplot(x = df_box['month'] , y = df_box['value'], ax = ax2)
    
    # Save image and return fig (don't change this part)
    fig.savefig('box_plot.png')
    return fig
```

```python
draw_line_plot()
```

```python
draw_bar_plot()
```

```python
draw_box_plot()
```
